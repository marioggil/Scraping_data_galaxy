```python
def otra_funcion():
    Obtener_nombres = driver.find_elements_by_xpath('//*[@class="participants-ul"]/li//span[@class="participants-item__display-name"]')
    lista_de_nombres = [x.text for x in Obtener_nombres]
    return lista_de_nombres


def obtener_numero():
    numero = driver.find_element_by_xpath('//*[@id="wc-container-right"]/div/div[1]/div[1]/div[2]/span')
    numero  = str(numero.text)
    numero = re.search("[0-9]+", numero)[0]
    numero= int(numero)
    return numero


def guardar_pandas(df, numero, lista, now, salio):
    df=df.append(pd.DataFrame([(numero,lista, now, salio)], columns=["Numero conectados", "Lista de nombres", "now","salio"]), ignore_index=True)
    return df
    

def numero_conectados(segundos):  
    df = pd.DataFrame(columns=["Numero conectados", "Lista de nombres", "now", "salio"])
    numero_1=2
    lista_de_nombres = otra_funcion()
    while True:
        salio=0
        try:
            numero= obtener_numero()
            if numero > numero_1:
                print("Entro alguien")
                lista_de_nombres = otra_funcion()
                numero_1 = numero
                salio=0
        
            elif numero < numero_1:
                print("Salio alguien")
                lista_de_nombres = otra_funcion() 
                numero_1 = numero
                salio=1
            print(lista_de_nombres)
            print(numero)
            now= time.strftime("%I:%M:%S")
            df = guardar_pandas(df, numero, lista_de_nombres, now, salio)
            time.sleep(segundos)
        except:
            return df
        
    return df
        
        
```
